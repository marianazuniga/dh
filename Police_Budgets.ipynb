# Police Budgets and Police Violence: Unpacking Social Reform in the Wake of the 21st century 


# Introduction

This project was inspired by one of my fundamental beliefs: the humanities exist to make the world a better place. This belief is being challenged in the present moment as demonstrations for systemic racial equity have taken over the globe following the racialized police killing of George Floyd. Although the momentum seen in the rallies is essential to galvanize mass support for reform, if systemic change is to be achieved, it will be achieved through concrete legislation that calls for a social restructuring. 

The purpose of this project is to assist those measures. Current tools such as the Urban Institute's State and Local Government Finance do not give itemized budget lists, or comparison to racialized killings within that city. The purpose of this analysis is to give context to current budget data in relation to other departments with similar racialized police killings. 


# Research Question

What is the relationship between police budget expenditures and racialized police violence in California? 


# Methodology

The Major Data Source I used for this project was Mapping Police Violences online tool (https://mappingpoliceviolence.org/cities). The tool is a comprehensive data set that houses data from 2013-2019, and allows comparison of racialized police killings, not only across police departments but across states.  
 

In order to gather the itemized police budgets, Adopted Fiscal Budgets were analyzed for fiscal years 2013/2014- fiscal year 2019/20 for the following city police departments: 
                Bakersfield, 
                Stockton, 
                Long Beach, 
                Santa Ana, 
                San Bernadino, 
                Fresno, 
                Anaheim,
                Fremont,
                Los Angeles, 
                Sacramento Police Departments.
